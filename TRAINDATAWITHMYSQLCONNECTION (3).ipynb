# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
import mysql.connector as sql
import pandas as pd

db_connection = sql.connect(host='localhost', database='aiml', user='root', password='root')
query = "select * from train"
df_mysql = pd.read_sql(query, con = db_connection)


survive = len(df_mysql[df_mysql["Survived"]==1])
nsurvive = len(df_mysql[df_mysql["Survived"]==0])
x = ['survived','not survived']
y = np.array([survive, nsurvive])

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(y, labels=x,autopct='%1.2f%%')
plt.show()
```
